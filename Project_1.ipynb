```python
# --- CELL 1: Imports and Configuration ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Set seed for reproducibility as per assignment requirements
RANDOM_STATE = 42

# Visual settings
%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python
# --- CELL 2: Data Loading and Cleaning ---

# Load data - low_memory=False handles mixed types in columns like 'budget'
movies_raw_data = pd.read_csv('dataset.csv', low_memory=False)

# Remove movies with no rating (Target variable must be present for training)
movies_dataset = movies_raw_data.dropna(subset=['averageRating']).copy()

# Simple numeric conversion: non-numbers will become NaN
movies_dataset['runtimeMinutes'] = pd.to_numeric(movies_dataset['runtimeMinutes'], errors='coerce')

# Fill missing/invalid runtimes with the median
movies_dataset['runtimeMinutes'] = movies_dataset['runtimeMinutes'].fillna(movies_dataset['runtimeMinutes'].median())

# Report the results
print(f"Initial movies: {len(movies_raw_data)}")
print(f"Final movies for modeling: {len(movies_dataset)}")
```

```python
movies_dataset.head()
```

```python
# --- CELL 3: Data Splitting ---

# 1. Define the target (the label we want to predict)
target_rating = movies_dataset['averageRating']

# 2. Define the features (the information used for prediction)
# We drop 'averageRating' so the model doesn't see the answer.
# We also drop 'tconst' and 'primaryTitle' from the features list 
# because they are IDs/Strings, but they still exist in 'movies_dataset'.
features_all = movies_dataset.drop(columns=['averageRating', 'tconst', 'primaryTitle'])

# 3. Split into Training (80%) and Testing (20%) sets
features_train, features_test, target_train, target_test = train_test_split(
    features_all, 
    target_rating, 
    test_size=0.2, 
    random_state=RANDOM_STATE
)

print(f"Split completed:")
print(f"- Training set: {features_train.shape[0]} movies")
print(f"- Testing set: {features_test.shape[0]} movies")
```
